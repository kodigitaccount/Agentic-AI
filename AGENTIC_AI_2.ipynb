```python
from transformers import GPTNeoForCausalLM, GPT2Tokenizer
import torch

# Load the pre-trained GPT-Neo model and tokenizer
model_name = "EleutherAI/gpt-neo-1.3B"
model = GPTNeoForCausalLM.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Move model to GPU if available, otherwise run on CPU
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Set the model to evaluation mode
model.eval()

# Function to generate a response based on user input
def generate_response(user_input):
    # Tokenize the input text
    input_ids = tokenizer.encode(user_input, return_tensors="pt").to(device)

    # Generate a response using the model
    output = model.generate(
        input_ids,
        max_length=150,  # Adjust for reasonable output length
        min_length=30,
        no_repeat_ngram_size=2,
        top_p=0.92,  # Nucleus sampling
        temperature=0.7,  # Controls randomness
        num_return_sequences=1,
        top_k=50,
        do_sample=True
    )

    # Decode and return the response
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response

# Interaction Loop
def chat_with_agent():
    print("Hello! I am your AI agent. Ask me anything. Type 'exit' to end the chat.")
    while True:
        # Get user input
        user_input = input("You: ")

        # Exit condition
        if user_input.lower() == "exit":
            print("Goodbye! See you next time.")
            break

        # Generate response using the model
        response = generate_response(user_input)
        print(f"AI Agent: {response}")

# Start the conversation
if __name__ == "__main__":
    chat_with_agent()
```
